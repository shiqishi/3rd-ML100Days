```python
from sklearn import datasets,linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
wine=datasets.load_wine()
boston=datasets.load_boston()
breast_cancer=datasets.load_breast_cancer()
```

```python
print(f'wine.data:\n{wine.data[:5]}')
print(f'wine.target:\n{wine.target}\n')
print(f'boston.data:\n{boston.data[:5]}')
print(f'boston.target:\n{boston.target}\n')
print(f'breast_cancer.data:\n{breast_cancer.data[:5]}')
print(f'breast_cancer.target:\n{breast_cancer.target}\n')
```

```python
x_train,x_test,y_train,y_test=train_test_split(wine.data,wine.target,test_size=0.2)
logreg=linear_model.LogisticRegression(multi_class='multinomial',solver='newton-cg')
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
acc=accuracy_score(y_test,y_pred)
print('Accuracy: ',acc)
```

```python
x_train,x_test,y_train,y_test=train_test_split(boston.data,boston.target,test_size=0.2)
lineReg=linear_model.LinearRegression()
lineReg.fit(x_train,y_train)
y_pred=lineReg.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
print('MSE: ',mse)
```

```python
x_train,x_test,y_train,y_test=train_test_split(breast_cancer.data,breast_cancer.target,test_size=0.2)
logreg=linear_model.LogisticRegression(solver='liblinear')
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
acc=accuracy_score(y_test,y_pred)
print('Accuracy: ',acc)
```
